```python
sns.pairplot(data=dfpop)
```

```python
sns.regplot(x=dfpop["POPULATION"], y=dfpop["incident_zip"], data=dfpop)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=dfpop["POPULATION"], y=dfpop["incident_zip"], data=dfpop, hue="descriptor") #hue means color the points by the certain column
plt.legend(loc=(1.05, 0.5))
plt.show()
```
