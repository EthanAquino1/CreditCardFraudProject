```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("Class", axis = 1)
y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=101)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.metrics import plot_confusion_matrix, classification_report
```

```python
log = LogisticRegression(max_iter=10000000)
tree = DecisionTreeClassifier()
forest = RandomForestClassifier()
ada = AdaBoostClassifier()
grad = GradientBoostingClassifier()
```

```python
def run_model(model, X_train, y_train, X_test, y_test):
  model.fit(X_train, y_train)
  model_preds = model.predict(X_test)
  print(classification_report(y_test, model_preds))
  print()
  plot_confusion_matrix(model, X_test, y_test)
  print()
```

```python
run_model(log, X_train, y_train, X_test, y_test)
```

```python
run_model(tree, X_train, y_train, X_test, y_test)
```

```python
run_model(forest, X_train, y_train, X_test, y_test)
```

```python
run_model(ada, X_train, y_train, X_test, y_test)
```

```python
run_model(grad, X_train, y_train, X_test, y_test)
```

```python
log_param_grid = {"penalty":["l1", "l2", "elasticnet"], 
                  "l1_ratio": np.linspace(0,1,20), 
                  "C": np.logspace(0,10,20)}
tree_param_grid = {"min_samples_split": [2,3,4,5], 
                   "max_depth": [2,3,4,5], 
                   "criterion":['gini', 'entropy']}
forest_param_grid = {"n_estimators": [64, 100, 128, 200], 
                     "max_features": [2, 3, 4],
                     "bootstrap": [True, False], 
                     "oob_score": [True, False]}
ada_param_grid = {"n_estimators": [50,100,150], 
                  "learning_rate": [0.1, 0.05, 0.2]}
grad_param_grid = {"n_estimators": [50,100,150], 
                    "learning_rate": [0.1, 0.05, 0.2],
                    "max_depth": [3,4,5]}
```

```python
log_grid_model = GridSearchCV(log, log_param_grid)
tree_grid_model = GridSearchCV(tree, tree_param_grid)
forest_grid_model = GridSearchCV(forest, forest_param_grid)
ada_grid_model = GridSearchCV(ada, ada_param_grid)
grad_grid_model = GridSearchCV(grad, grad_param_grid)
```

```python
log = LogisticRegression(C = 10, penalty = "l2", max_iter=10000000)
tree = DecisionTreeClassifier(max_depth=5, min_samples_split=2, criterion="gini")
forest = RandomForestClassifier(n_estimators=128, max_features=4, bootstrap=False, oob_score = False)
ada = AdaBoostClassifier(learning_rate=0.2, n_estimators=150)
grad = GradientBoostingClassifier(learning_rate=0.2, n_estimators=150, max_depth=3)
```
